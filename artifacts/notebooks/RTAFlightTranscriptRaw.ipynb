```python
import requests
from azure.eventhub import EventHubProducerClient, EventData
```

```python
eventstream=''
producer = EventHubProducerClient.from_connection_string(conn_str=eventstream)

i=0

for i in range(0, 436554):
    df = requests.get('https://loripsum.net/api/6000/short/plaintext')
    b = producer.create_batch() 
    j = df.content
    # print(j)
    b.add(EventData(j))
    producer.send_batch(b) # send it!
    # time.sleep(1)
    producer.close()
    i=i+1    
```
